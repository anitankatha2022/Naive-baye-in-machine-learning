```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('C:\\Users\\USER\\Desktop\\Toronto\\autism_screening.csv')
df.head()
```

```python
df.drop(['contry_of_res','used_app_before','relation','age_desc','Class/ASD','ethnicity','jundice','austim','result'],axis='columns',inplace=True)
df.head()    
```

```python
target=df.age
input=df.drop('A4_Score',axis='columns')
```

```python
dummies=pd.get_dummies(input.gender)
dummies.head()
```

```python
input=pd.concat([input,dummies],axis='columns')
input.head()
```

```python
input.drop('gender',axis='columns',inplace=True)
input.head()
```

```python
input.columns[input.isna().any()]
```

```python
df.dropna(inplace=True)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(input,target,test_size=0.3)
```
